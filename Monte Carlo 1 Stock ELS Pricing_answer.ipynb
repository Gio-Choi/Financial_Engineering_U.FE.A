```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)

S=100# 기초자산
T=3 # 3년
r=0.01 
std=0.2 # vol
Nsimulations=1000 #1000 시뮬레이션
days=252 
Nsteps=days*T 
K=100 # 행사가
dt = 1/Nsteps 
Sample_Stock = np.zeros((Nsimulations,Nsteps))
```

```python
pd.DataFrame(Sample_Stock)
```

```python
drift = (r-(std**2)/2)*dt
a = std*np.sqrt(dt)
x = np.random.normal(0,1,(Nsimulations,Nsteps))
```

```python
pd.DataFrame(x)
```

```python
for i in range(0,Nsimulations):
    for j in range(1, Nsteps):
        Sample_Stock[i,0] = S
        Sample_Stock[i,j] = Sample_Stock[i,j-1]*np.exp(drift+a*x[i,j-1])
```

```python
df1=pd.DataFrame(Sample_Stock)
df1=df1.transpose() # 행 열 바꾸기
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(df1)
```

```python
df1.columns.name="simulations"
df1.index.name="steps"
df1
```

```python
profit = np.zeros((8,Nsimulations))
r2=0.06
time1=Nsteps*1/6 # 6개월
time2=Nsteps*2/6 # 12개월
time3=Nsteps*3/6 # 18개월
time4=Nsteps*4/6 # 24개월
time5=Nsteps*5/6 # 30개월
time6=Nsteps*6/6 # 36개월
```

```python
pd.DataFrame(profit)
```

```python
for i in range(0,Nsimulations):
    if df1.loc[time1-1,i]>S*0.90: # 6개월 시점에서 초기 가격 S(100)의 90% 보다 큰 값이 나온다면 조기상환
        profit[0,i]=S*(1+r2)**0.5*np.exp(-r*0.5)
        
    elif df1.loc[time2-1,i]>S*0.90: # 12개월 시점에서 초기 가격 S(100)의 90% 보다 큰 값이 나온다면 조기상환
        profit[1,i]=S*(1+r2)**1*np.exp(-r*1)
        
    elif df1.loc[time3-1,i]>S*0.80: # 18개월 시점에서 초기 가격 S(100)의 80% 보다 큰 값이 나온다면 조기상환
        profit[2,i]=S*(1+r2)**1.5*np.exp(-r*1.5)
        
    elif df1.loc[time4-1,i]>S*0.70: # 24개월 시점에서 초기 가격 S(100)의 70% 보다 큰 값이 나온다면 조기상환
        profit[3,i]=S*(1+r2)**2*np.exp(-r*2)
        
    elif df1.loc[time5-1,i]>S*0.60: # 30개월 시점에서 초기 가격 S(100)의 60% 보다 큰 값이 나온다면 조기상환
        profit[4,i]=S*(1+r2)**2*np.exp(-r*2.5)
        
    elif df1.loc[time6-1,i]>S*0.50: # 36개월 시점에서 초기 가격 S(100)의 50% 보다 큰 값이 나온다면 조기상환
        profit[5,i]=S*(1+r2)**2*np.exp(-r*3)
        
    elif min(df1.loc[:,i])> S*0.40: # 36개월 시점에서 초기 가격 S(100)의 50% 보다 작은 값이지만 한번도 40% 밑으로 떨어진 적이 없다면
        profit[6,i]=S*np.exp(-r*3)  # 수익률 없이 원금 S
        
    elif min(df1.loc[:,i])< S*0.40: # 36개월 시점에서 초기 가격 S(100)의 50% 보다 작은 값이고 40% 밑으로 떨어진 적이 있을 때
        profit[7,i]=S*np.exp(-r*3)  # 원금 * (만기가격/원금) 이므로 그냥 만기가격 가져감 <- 2stock 일 때는 다름
```

```python
df2=pd.DataFrame(profit)
df2.columns.name="simulations"
df2.index.name="case"
df2
```
